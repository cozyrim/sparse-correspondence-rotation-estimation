```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np

def plot_point_on_image(image_path, coordinates):
    # 이미지 로드
    img = Image.open(image_path)
    fig, ax = plt.subplots()
    ax.imshow(img)
    
    # 좌표에 원 그리기
    for (x, y) in coordinates:
        circle = patches.Circle((x, y), radius=10, edgecolor='red', facecolor='none')
        ax.add_patch(circle)
    
    # 축 제거
    ax.axis('off')
    plt.show()

# 사용 예시
image_path = "C:/Users/chaerim/OneDrive/Calibration/matching/cam1_cam2_2degree_split_c/left/img_1_00027.jpg"  # 이미지 경로 지정
coordinates = [575, 43], [593, 43], [612, 42], [629, 43], [652, 42],[661, 52], [640, 52], [649, 61], [657, 71], [619, 52],[628, 60], [638, 71], [646, 81], [655, 91], [602, 52],[609, 61], [617, 70], [626, 78], [635, 88], [644, 98]  # 원을 그릴 좌표 지정
plot_point_on_image(image_path, coordinates)
```

```python
import cv2
import matplotlib.pyplot as plt

# 이미지 파일 경로 (사용자의 이미지 파일 경로로 대체)
image1_path = 'C:/Users/chaerim/OneDrive/Calibration/matching/matching/images/dog_edges_L13.png'
image2_path = 'C:/Users/chaerim/OneDrive/Calibration/matching/matching/images/dog_edges_R00.png'

# 이미지를 불러오기
image1 = cv2.imread(image1_path)
image2 = cv2.imread(image2_path)

# BGR을 RGB로 변환 (Matplotlib에서 올바르게 표시되도록)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# 이미지 1과 2의 대응점들
points_image1 = [(622, 60), (590, 107), (629, 130), (660, 83), (663, 117), (666, 148), 
                 (696, 106), (637, 192), (608, 230), (646, 246), (672, 208), (676, 235), 
                 (692, 245), (690, 218), (658, 277), (628, 288), (652, 301), (702, 278), 
                 (668, 290), (698, 303)]

points_image2 = [(212, 79), (185, 121), (222, 148), (248, 109), (253, 139), (256, 168), 
                 (279, 134), (232, 205), (207, 242), (240, 257), (263, 221), (266, 247), 
                 (281, 254), (278, 230), (252, 284), (225, 296), (244, 307), (287, 284), 
                 (260, 296), (283, 306)]

# 이미지에 점을 표시하는 함수
def plot_points(image, points, color='red'):
    for (x, y) in points:
        cv2.circle(image, (x, y), 5, (255, 0, 0), -1)  # 점의 색상: 빨강

# 이미지에 대응점을 표시
plot_points(image1, points_image1)
plot_points(image2, points_image2)

# 이미지 출력
fig, axes = plt.subplots(1, 2, figsize=(15, 10))
axes[0].imshow(image1)
axes[0].set_title("Image 1 with Keypoints")
axes[0].axis("off")

axes[1].imshow(image2)
axes[1].set_title("Image 2 with Keypoints")
axes[1].axis("off")

plt.show()
```
